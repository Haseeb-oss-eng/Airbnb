```python
#Simple imputer
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='most_frequent')
df['property_type'] = imputer.fit_transform(df[['property_type']]).ravel()
```

```python
df.isna().aggregate('sum')
```

####Summary
After preprocessing the datasets using SimpleImputer using most-frequent strategy property_type is filled.

```python
#remove the dollar sign from the price column
df['price_F'] = df['price'].replace('[\$,]','',regex=True).astype(float)
df['extra_people_F'] = df['extra_people'].replace('[\$,]','',regex=True).astype(float)
```

```python
df['extra_people_F'].head(3)
```

```python
def amenity_count(data):
  splitted = data.split(',')
  length = len(splitted)
  return length
```

```python
df['amenities_count'] = df['amenities'].apply(amenity_count)
```

```python
import re
import ast

def split_amenities(data):
  splitted = data.replace('{','[').replace('}',']')
  quoted = re.sub(r'(?<=\[|,)(\s*[^,"\[\]]+?\(?.*?\)?)(?=,|\])', r'"\1"', splitted)
  listed = ast.literal_eval(quoted)
  return listed
```

```python
df['amenities'] = df['amenities'].apply(split_amenities)
df
```

```python
from geopy.distance import geodesic

def distance(data):
  for i in range(len(data) - 1):
        point1 = (data.loc[i, 'latitude'], data.loc[i, 'longitude'])
        point2 = (data.loc[i+1, 'latitude'], data.loc[i+1, 'longitude'])
        dist_km = geodesic(point1, point2).km
        data.loc[i+1, 'distance'] = dist_km
  return data
```
